```python
import plotly.express as px
import pandas as pd

df = pd.read_csv('data2.csv')

df
```

```python
column_names = ['大專院校', '高級中學', '職業學校', '國民中學','國民小學']
fig = px.line(df, x="學年度別", y=column_names,
              hover_data={"學年度別", "學年度別"},
              title='70-110各級學校每生擁有經費')
fig.update_xaxes(
    dtick="M1",
    tickformat="%b\n%Y",
    ticklabelmode="period")
fig.show()
```

#### Q1:請問最新的年度中各級學校每生擁有經費依序為何?
#### A: 國中 => 職校 => 大專院校 => 高中 => 國小

#### Q2:圖中可以看到原本大專院校一直都是經費最多的，但後續被國中及職校超越，請問分別是在幾年?可能原因為何?
#### A: 國中在106~107年度間超越大專院校，職校於108~109年度間超越大專院校，推測政府因推動108課綱而增加經費

#### Q3:大專院校在哪年擁有歷史最高經費?可能原因為何?
#### A: 96年度，推測因開始推動大學多元入學方案，而提高大專院校經費
