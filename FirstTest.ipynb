```python
import pandas as pd

df = pd.read_csv('data2.csv')

df.head()
```

```python
column_names = ['大專院校', '高級中學', '職業學校', '國民中學','國民小學']
data = {
    "總經費": df[column_names].sum(),
    "平均每年經費": df[column_names].sum()/41,
}
Type = pd.DataFrame(data)
Type['統計年分'] = '民國70年~110年'
Type
```

```python
data2 = {
    "總經費": df.iloc[19:41][['幼兒園']].astype(float).sum(),
    "平均每年經費":df.iloc[19:41][['幼兒園']].astype(float).sum()/22
}
Type2 = pd.DataFrame(data2)
Type2['統計年分'] = '民國89年~110年'
Type2
```

```python
frames = [Type, Type2]

result = pd.concat(frames)
result
```
