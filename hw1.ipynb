## 臺北市市立各級學校每生享有教育經費

```python
import pandas as pd

df = pd.read_csv('data2.csv')

df
```

### Q1.各級學校每生經費最多的分別是在民國幾年?多少經費?(最多->最少)

```python
column_names = ['大專院校', '高級中學', '職業學校', '國民中學','國民小學']
list = []
year = []

list = df.loc[:,column_names].astype(float).max()
year = df.loc[:,column_names].astype(float).idxmax()+ 70 


data = {
    "最高經費": list,
    "年分" : year.astype(str) + '年'
    
}
data2 = {
    "最高經費": df.iloc[19:41][['幼兒園']].astype(float).max(),
    "年分" : (df.iloc[19:41][['幼兒園']].astype(float).idxmax() + 70).astype(str) + '年'
}
Type = pd.DataFrame(data)
Type2 = pd.DataFrame(data2)

frames = [Type, Type2]
result = pd.concat(frames)
result1 = result
result 
Type = result.sort_values(by=['最高經費'],ascending=False)
Type = Type[['最高經費','年分']]
Type
```

```python
column_names = ['大專院校', '高級中學', '職業學校', '國民中學','國民小學']
list = []
year = []

list = df.loc[:,column_names].astype(float).min()
year = df.loc[:,column_names].astype(float).idxmin()+ 70 


data = {
    "最低經費": list,
    "年分" : year.astype(str) + '年'
    
}
data2 = {
    "最低經費": df.iloc[19:41][['幼兒園']].astype(float).min(),
    "年分" : (df.iloc[19:41][['幼兒園']].astype(float).idxmin() + 70).astype(str) + '年'
}
Type = pd.DataFrame(data)
Type2 = pd.DataFrame(data2)

frames = [Type, Type2]
result = pd.concat(frames)
result2 = result
result
Type = result.sort_values(by=['最低經費'],ascending=True)
Type = Type[['最低經費','年分']]
Type
```

### Q5.各級學校每生歷年總經費及每年總經費平均為何?

```python
column_names = ['大專院校', '高級中學', '職業學校', '國民中學','國民小學']

data = {
    "總經費": df[column_names].sum(),
    "平均每年經費": df[column_names].sum()/41,
    "統計年分": "70年~110年"
    }
data2 = {
    "總經費": df.iloc[19:41][['幼兒園']].astype(float).sum(),
    "平均每年經費": df.iloc[19:41][['幼兒園']].astype(float).sum()/23,
    "統計年分": "89年~110年"
}
Type = pd.DataFrame(data)
Type2 = pd.DataFrame(data2)
frames = [Type, Type2]
result = pd.concat(frames)
result
```

### Q6.各級學校每生每年總經費平均排序為何?
